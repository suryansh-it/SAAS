```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data from a CSV file
df = pd.read_csv('C:\\Users\\soura\\Downloads\\Compressed\\9 Smart-Attendence-System (Working)-20240405T174651Z-001\\9 Smart-Attendence-System (Working)\\bigger_attendance.csv')
```

```python
df.describe()
```

```python
df['Time'] = pd.to_datetime(df['Time'])

# Calculate time in minutes
df['Time_minutes'] = df['Time'].dt.hour * 60 + df['Time'].dt.minute + df['Time'].dt.second / 60

# Check if 'Time_minutes' column is added correctly
print(df.head())  # Optional: Print the first few rows of the DataFrame to verify the column

# Create a boxplot
plt.figure(figsize=(8, 6))
plt.boxplot(df['Time_minutes'])
plt.title('Boxplot of Attendance Times')
plt.ylabel('Time (minutes)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(df['Time'], labels=df['Name'], autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Time')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Name'], df['Time'])
plt.title('Bar Plot')
plt.xlabel('Name')
plt.ylabel('Time')
plt.show()
```
